# Imports

```python
from diffusers import StableDiffusionPipeline
from utils import image_grid
from hashlib import md5
import torch
```

# Parameters

```python
model_id = "stabilityai/stable-diffusion-2-1"
inf_steps = 20
no_imgs = 200
cols = no_imgs // 2 if no_imgs // 2 <= 5 else 5
rows = no_imgs // cols
img_output_folder = "./models/shane/data/man"
file_extension = ".jpg"
```

# Model

```python
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")
```

```python
prompt = "photo of a young man in colour, realistic, photography, hard rim lighting photography--beta --ar 2:3  --beta --upbeta"
n_prompt = "bad, deformed, ugly, bad anatomy"
image = [0] * no_imgs
for i in range(no_imgs):
    image[i] = pipe(prompt, negative_prompt=n_prompt, num_inference_steps=50).images[0]
```

```python
#image_grid(image, rows=rows, cols=cols)
```

```python
for idx, img in enumerate(image):
    img_path = img_output_folder + "/" + str(idx) + "_" + md5(img.tobytes()).hexdigest() + file_extension
    img.save(img_path)
```
